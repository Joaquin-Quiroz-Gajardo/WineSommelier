# Become a sommelier

```python
%matplotlib inline
import pandas as pd
import nltk
import re
import numpy as np

from nltk import word_tokenize
from nltk.stem import WordNetLemmatizer 
from nltk.corpus import stopwords
from string import punctuation
from functools import reduce
from sklearn.model_selection import train_test_split
import pdb

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
filename = '../DataBase/5_grape_db.xlsx'

a0 = pd.read_excel(filename)

columns_sel = ['abv', 'colour', 'country', 'description', 'grape_variety', 'name', 'Body']

grapes = a0['grape_variety'].tolist()

# in case we do not want to blend grapes
grape = [grape for x in grapes if len(x.split(','))==1 for grape in x.split(', ')]
fdist = nltk.FreqDist(grape)
```

```python
fdist.plot(20)
```

```python
result = a0['grape_variety']

## removing varieties that have only one member in the database
counts = nltk.Counter(result)
varieties = [key for key in counts if counts[key] > 30]
data_input = a0[a0['grape_variety'].isin(varieties)].reset_index()
```

```python
# split the data into train and test
combined_features = ['Body', 'description', 'grape_variety']
target = 'grape_variety'

X_train, X_test, y_train, y_test = train_test_split(data_input[combined_features], data_input[target], test_size=0.33, random_state=42)
```

```python
ax = sns.barplot(x='grape_variety', y='Body', data=X_train.groupby('grape_variety', as_index=False).count())
ax.set_xticklabels(ax.get_xticklabels(), rotation=60)
ax.set_ylabel('Count')
```

```python
# aggregate description by grape type
grouped = X_train[['grape_variety', 'description']].groupby(['grape_variety']).agg({'description': lambda z: reduce(lambda x,y: ''.join(x+y), z)})

# tokenize the combined text
grouped['tokens'] = grouped.loc[:, 'description'].apply(word_tokenize)
```

```python
grouped.head()
```

```python
def word_count_df(df, src_col, out_col=('grape', 'token', 'count')):
    dfp = pd.DataFrame()
    for ii in df.index:
        dist = nltk.FreqDist(df.loc[ii, src_col])
        dftmp = pd.DataFrame(dict(zip(out_col, ([ii]*len(dist.keys()), list(dist.keys()), list(dist.values())))))
        dfp=pd.concat([dfp, dftmp], ignore_index=True)
    return(dfp)

tkn_count = word_count_df(grouped, 'tokens', out_col=('grape', 'token', 'count'))
```

```python
tkn_count['rank']=tkn_count.groupby(['grape'], as_index=False)['count'].rank(ascending=False)
tkn_count.head()
```

```python
# plot the word count of each grape type
def barplot_wordcounts(df, types, m=2, n=2):
    sns.set(font_scale=1.3)
    fig, axarr = plt.subplots(m, n, figsize=(13, 6))
    plt.subplots_adjust(top=0.99, bottom=0.15, left=0.10, right=0.95, hspace=0.8,
                    wspace=0.25)
    ii = 0
    for row in axarr:
        for col in row:
            if ii < len(types):
                col.set_title(types[ii])
                filt_cond = (df['rank'] <= 15) & (df['grape']==types[ii])
                sns.barplot(x='token', y='count', ci = None,
                    data=df[filt_cond].sort_values('count', ascending=False), 
                    ax = col)
                ii += 1
                col.set_xticklabels(col.get_xticklabels(), rotation=60)
                col.set_xlabel('')

barplot_wordcounts(tkn_count, grouped.index, m=3)
```

As we can see the most frequent 15 words are basically filling words or some kind of punctuations. Except for the word 'citrus' (sauvignon blanc) and 'red' (pinot noir) there are no other words in the plot that may give us a clue about the variety of the grape. Therefore it makes sense to filter the text from these filling words that make noise in the text and does not help to determine the variety of the grape. 

In the first round we will apply a simple non alphabetic character remover to get rid of words with length of 1 and 2 characters, full stops, commas, numbers, etc. Then we apply a stopword filter from the nltk.stopwords package appended by some additional words that do not contain any usefool information about the variety of grapes like the word 'wine' and some more.

```python
pattern = re.compile(r'[0-9]+|\b[\w]{2,2}\b|[%.,_`!"&?\')({~@;:#}+-]+|\b[\w]{1,1}\b')

tkn_count['token'] = tkn_count['token'].apply(lambda t: pattern.sub('', t))
tkn_filt = tkn_count[-tkn_count['token'].isin([''])].reset_index()


stop_words = stopwords.words('english')
stop_append = ['.', ',', '`', '"', "'", '!', ';', 'wine', 'fruit', '%', 'flavours', 'aromas', 'palate']
stop_words = stop_words + stop_append

tkn_filt1 = tkn_filt[-tkn_filt['token'].isin(stop_words)].reset_index()
tkn_filt1['rank'] = tkn_filt1.groupby(['grape'], as_index=True)['count'].rank(ascending=False)
tkn_filt1 = tkn_filt1.drop(['level_0', 'index'], axis=1)

```

```python
barplot_wordcounts(tkn_filt1, grouped.index, m=3)
```

These steps allow us to remove some basic noise from the collection of words but the remaining collection of words is still not clean enough. Particularly, some words may appear in different forms while actually having the same meaning. For example, in the case of pinot noir we can see the words cherry and cherries. They are the same but the former is singular while the latter is plural. To resolve this phenomena we can apply lemmatization. This process tries to group words into their common form. 

```python
def is_noun(tag):
    return tag in ['NN', 'NNS', 'NNP', 'NNPS']

def is_verb(tag):
    return tag in ['VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ']

def is_adverb(tag):
    return tag in ['RB', 'RBR', 'RBS']

def is_adjective(tag):
    return tag in ['JJ', 'JJR', 'JJS']

def penn_to_wn(tag):
    if is_adjective(tag):
        return nltk.stem.wordnet.wordnet.ADJ
    elif is_noun(tag):
        return nltk.stem.wordnet.wordnet.NOUN
    elif is_adverb(tag):
        return nltk.stem.wordnet.wordnet.ADV
    elif is_verb(tag):
        return nltk.stem.wordnet.wordnet.VERB
    return nltk.stem.wordnet.wordnet.NOUN
    
```

```python
lemm = WordNetLemmatizer
ptags = nltk.pos_tag(tkn_filt1['token'].tolist())
ptags_wn = [(t[0], penn_to_wn(t[1])) for t in ptags]
lems = [lemm.lemmatize(tag[1], tag[0]) for tag in ptags_wn]
tkn_filt1['token'] = lems
tkn_filt1.shape
```

After filtering, some of the taste and smell characteristics start to apppear in the most frequent words. These words differentiate the varieties. There are still some words that do not bear any information about the difference between the grapes, but for the moment we leave them in.

Let us now investigate what are the common and disjoint words for these grapes. 

```python
tkn_filt1 = tkn_filt1.groupby(['grape', 'token'], as_index=False)['count'].sum()
```

```python
tkn_filt1['rank'] = tkn_filt1.groupby(['grape'], as_index=True)['count'].rank(ascending=False)
barplot_wordcounts(tkn_filt1, grouped.index, m=3)
```

```python
dfs = []
for gr in varieties:
    tmp = tkn_filt1[tkn_filt1.grape == gr]
    tmp = tmp.add_suffix('_'+gr)
    tmp.columns=tmp.columns=tmp.columns.str.replace('token_'+gr,'token')
    dfs.append(tmp)
    

df_final = reduce(lambda left,right: pd.merge(left,right,on='token', how='outer'), dfs)
```

```python
df_common = df_final.dropna()
cols = df_common.columns.str.contains('count') | df_common.columns.str.contains('token')
df_common = df_common[df_common.columns[cols]]
df_common.corr()

```

Check what are the most common words

```python
from scipy.stats import zscore
sort_col = df_common.columns.str.contains('count')
df_common['freq'] = df_common[df_common.columns[sort_col]].apply(zscore).sum(axis=1)
#ind = df_common[df_common.columns[sort_col]].apply(zscore).sum(axis=1).sort_values(ascending=False).index
ind = df_common['freq'].sort_values(ascending=False).index
df_common.loc[ind[:10], ]
#sns.relplot(x=list(range(10)), y=[0]*10, size='freq', data=df_common)
```

```python
from matplotlib import cm

cmap = cm.get_cmap('cool')

dfplot = df_common.sort_values(by='freq',ascending=False)
dfplot.index = dfplot['token']
sort_col = df_common.columns.str.contains('count')
dfplot['csum'] = dfplot[dfplot.columns[sort_col]].sum(axis=1)
sort_col = dfplot.columns[dfplot.columns.str.contains('count')]
llabel = []
for col in sort_col:
    llabel.append(col.replace('count_',''))
    dfplot[col] = dfplot[col] / dfplot['csum']

ax = dfplot.loc[list(reversed(dfplot.index[:10])), sort_col].plot(kind='barh', stacked=True, cmap=cmap)
ax.legend(llabel, bbox_to_anchor=(1., 1))
ax.set_facecolor('w')
```

```python
df_unique = pd.DataFrame()
for gr in varieties:
    cond = ~(df_final.columns.str.contains(gr) | df_final.columns.str.contains('token'))
    ind = df_final[df_final.columns[cond]].isna().all(1)
    tmp = df_final.loc[ind, ~cond]
    tmp.columns = tmp.columns.str.replace('_'+gr, '')
    df_unique = pd.concat([df_unique, tmp], ignore_index=True, sort = True)
    
    
```

```python
df_unique['rank'] = df_unique.groupby(['grape'], as_index=False)['count'].rank(ascending=False)
barplot_wordcounts(df_unique, varieties, m=3)



```
